```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"
```

```python
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
```

```python
combined_df = pd.merge(ride_df,city_df,on='city')
```

```python
#Urban City

urban_city_df = combined_df.loc[combined_df['type'] == 'Urban']

urban_total_fare = urban_city_df['fare'].sum()
urban_total_rides = urban_city_df['ride_id'].count()

urban_citylist_df = urban_city_df.groupby('city')
urban_driver_count = urban_citylist_df['city','driver_count'].mean()
urban_total_drivers = urban_driver_count.sum()

number_of_rides_urban = urban_citylist_df['city'].count()

average_fare_urban = urban_citylist_df['city','fare'].mean()
```

```python
#Rural City

rural_city_df = combined_df.loc[combined_df['type'] == 'Rural']

rural_total_fare = rural_city_df['fare'].sum()
rural_total_rides = rural_city_df['ride_id'].count()

rural_citylist_df = rural_city_df.groupby('city')
rural_driver_count = rural_citylist_df['city','driver_count'].mean()
rural_total_drivers = rural_driver_count.sum()

number_of_rides_rural = rural_citylist_df['city'].count()

average_fare_rural = rural_citylist_df['city','fare'].mean()
```

```python
#Suburbal City

suburban_city_df = combined_df.loc[combined_df['type'] == 'Suburban']

suburban_total_fare = suburban_city_df['fare'].sum()
suburban_total_rides = suburban_city_df['ride_id'].count()

suburban_citylist_df = suburban_city_df.groupby('city')
suburban_driver_count = suburban_citylist_df['city','driver_count'].mean()
suburban_total_drivers = suburban_driver_count.sum()

number_of_rides_suburban = suburban_citylist_df['city'].count()

average_fare_suburban = suburban_citylist_df['city','fare'].mean()
```

```python
#% of Total Fare by City Type

labels = ["Urban", "Rural", "Suburban"]
sizes = [urban_total_fare, rural_total_fare, suburban_total_fare]
colors = ["lightcoral", "yellow", "lightskyblue"]
explode = (0.1,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=300)
plt.title('% of Total Fare by City Type')
plt.show()
```

```python
#% of Total Rides by City Type

labels = ["Urban", "Rural", "Suburban"]
sizes = [urban_total_rides, rural_total_rides, suburban_total_rides]
colors = ["lightcoral", "yellow", "lightskyblue"]
explode = (0.1,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=300)
plt.title('% of Total Rides by City Type')
plt.show()
```

```python
#% of Total Drivers by City Type

labels = ["Urban", "Rural", "Suburban"]
sizes = [urban_total_drivers, rural_total_drivers, suburban_total_drivers]
colors = ["lightcoral", "yellow", "lightskyblue"]
explode = (0.1,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=300)
plt.title('% of Total Drivers by City Type')
plt.show()
```

```python
plt.scatter(number_of_rides_urban, average_fare_urban, marker="o", facecolors="coral", edgecolors="black",
                            s=10*urban_driver_count, alpha=0.6, label="Urban")

plt.scatter(number_of_rides_rural, average_fare_rural, marker="o", facecolors="gold", edgecolors="black", 
                            s=10*rural_driver_count, alpha=0.6, label="Rural")

plt.scatter(number_of_rides_suburban, average_fare_suburban, marker="o", facecolors="lightskyblue", edgecolors="black", 
                            s=10*suburban_driver_count, alpha=0.6, label="Suburban")

plt.legend(loc= "upper right", title= "City Types")
plt.grid()
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.show()
```
